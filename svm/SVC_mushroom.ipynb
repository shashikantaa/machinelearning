```python
x=df.drop(['class'],axis=1)
x
```

```python
y=df['class']
y
```

```python
le=LabelEncoder()
x=x.apply(le.fit_transform)
x
```

```python
ml=svm.SVC()
```

```python
ml.fit(x_train,y_train)
```
