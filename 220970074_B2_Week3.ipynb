```python


#Exercise 2

import pandas as pd
import seaborn as sns
import numpy as np

# Load the dataset
data = pd.read_csv(r"C:\Users\shash\OneDrive\Desktop\mit\3rd sem\mca\220970074_B2_Week2.csv")

# (i) Retrieve the column names
column_names = data.columns
print("Column Names:")
print(column_names)

# (ii) Handling missing/Null values
# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values:")
print(missing_values)

# You can handle missing values based on your specific use case, e.g., by imputing them.

# (iii) Draw heatmaps for numerical columns
# Filter out numerical columns
numerical_columns = data.select_dtypes(include=[np.number])

# Create a correlation matrix
correlation_matrix = numerical_columns.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Determine related columns based on correlation values
related_columns = correlation_matrix.unstack().sort_values(ascending=False)
print("\nRelated Columns (Highest Correlations):")
print(related_columns)

# (iv) Calculate Spearman rank correlation coefficients and covariance
spearman_corr = numerical_columns.corr(method='spearman')
covariance_matrix = numerical_columns.cov()

print("\nSpearman Rank Correlation Coefficients:")
print(spearman_corr)

print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
#Exercise 5

#2) For the .csv file created in the above question, draw the following plots:
#    (i) Box plots for all numerical columns {M1  M5}
#   (ii) And also draw all possible charts for numerical columns.


import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\shash\OneDrive\Desktop\mit\3rd sem\mca\220970074_B2_Week2.csv")

# (i) Box plots for all numerical columns {M1  M5}
numerical_columns = ['M1', 'M2', 'M3', 'M4', 'M5']

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numerical_columns])
plt.title('Box Plots for Numerical Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.show()

# (ii) Create all possible charts for numerical columns
# Pairplot (Scatter plots and histograms for all numerical columns)
sns.pairplot(df[numerical_columns])
plt.suptitle('Pairplot for Numerical Columns', y=1.02)
plt.show()

# Histograms for each numerical column
for col in numerical_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram for {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Scatter plots between all numerical columns
sns.pairplot(df[numerical_columns], kind='scatter')
plt.suptitle('Scatter Plots for Numerical Columns', y=1.02)
plt.show()

# You can create additional types of plots as needed.
```
