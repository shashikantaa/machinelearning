```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rand
```

```python
mu = np.array([10,13])
sigma= np.array([[3.5,-1.8],[-1.8,3.5]])
print("mu",mu.shape)
print("sigma",sigma.shape)
```

```python
org_data=rand.multivariate_normal(mu,sigma,size=(1000))
print("Data shape",org_data.shape)
type(org_data)
```

```python
org_data[0]
```

```python

x1=org_data[:,0]
x1.shape
```

```python
x1[0]
```

```python
y1=org_data[:,1]
y1.shape
```

```python
y1[0]
```

```python
plt.scatter(x1,y1,c='b')
```

# step 2:- mean centering/normalize data

```python

#substract mean from data
mean = np.mean(org_data,axis=0)
mean
```

```python
type(mean)
```

```python
print('mean',mean.shape)
```

```python
meanx=np.mean(x1,axis=0)
meanx
```

```python
meany=np.mean(y1,axis=0)
meany
```

```python
x1-meanx
```

```python
y1-meany
```

```python
mean_data=org_data-mean
mean_data
```

```python
mean_data.shape
```

```python
plt.subplot(1,2,1)
plt.scatter(x1,y1,c='b')
```

```python
mean_data[0]
```

```python
x1mean=mean_data[:,0]
x1mean.shape
```

```python
y1mean=mean_data[:,1]
y1mean.shape
```

```python
plt.subplot(1,2,2)
plt.scatter(x1mean,y1mean,c='r')
```

```python
cov= np.cov(mean_data.T)
cov= np.round(cov,2)
print("covariance matrix",cov.shape,"\n")
```

```python
print(cov)
```

```python
# perform eigen decomposition of covariance matrix
eig_val,eig_vec=np.linalg.eig(cov)
print("Eigen vectors",eig_vec)
print("\nEgen values",eig_val,"\n")
```
