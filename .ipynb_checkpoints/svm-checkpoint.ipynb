```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv(r"D:\mca_074\mca_074ml\OneDrive_1_15-8-2023\DATASETS-16AUG2023\play_tennis.csv")
```

```python
df1
```

```python
df=df1.drop(columns='day')
df
```

```python
df['temp']=df['temp'].astype('category')
df['temp']=df['temp'].cat.codes
df
```

```python
df['outlook']=df['outlook'].astype('category')
df['outlook']=df['outlook'].cat.codes
df
```

```python
df['humidity']=df['humidity'].astype('category')
df['humidity']=df['humidity'].cat.codes
df
```

```python
df['wind']=df['wind'].astype('category')
df['wind']=df['wind'].cat.codes
df
```

```python
df['play']=df['play'].astype('category')
df['play']=df['play'].cat.codes
df
```

```python
x=df.drop(['play'],axis=1).values
x
```

```python
y=df["play"].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
clf=svm.SVC()
```

```python
clf.fit(x_train,y_train)
```

```python
clf.predict(x_test)
```

```python
clf.support_vectors_
```

```python
clf.support_
```

```python
clf.n_support_
```
