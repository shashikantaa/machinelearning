```python
le = LabelEncoder()
data_coloumns = ['Formatted Date' , 'Summary' , 'Precip Type' , 'Daily Summary']
for columns in data_coloumns:
    df[columns] = le.fit_transform(df[columns])
df.head() 
```

```python
#function for covariance matrix

def cova(df):
    
    #converting the dataframe into array for calculation
    data_array = df.to_numpy()
    
    #calculating mean
    mean = np.mean(data_array , axis = 0)
    
    #calculating the center point
    center = data_array - mean
    
    #calculating the covariance by multiplying it 
    cov_mat = np.dot(center.T , center) / (len(data_array) - 1)
    
    return cov_mat

cov_m = cova(df)
cov_m
```

```python
#getting the eigen values and eigen vectors
eigen_values , eigen_vectors = np.linalg.eig(cov_m)
print(eigen_values)
print(eigen_vectors)
```
